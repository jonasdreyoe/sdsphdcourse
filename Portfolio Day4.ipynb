Take the following text and transform it into a list of lists with with each element being a tokenized sentence. Remove stopwords, lower all tokens and keep only alpha-numeric tokens.


"I’ve been called many things in my life, but never an optimist. That was fine by me. I believed pessimists lived in a constant state of pleasant surprise: if you always expected the worst, things generally turned out better than you imagined. The only real problem with pessimism, I figured, was that too much of it could accidentally turn you into an optimist."


```
text = "I’ve been called many things in my life, but never an optimist. That was fine by me. I believed pessimists lived in a constant state of pleasant surprise: if you always expected the worst, things generally turned out better than you imagined. The only real problem with pessimism, I figured, was that too much of it could accidentally turn you into an optimist."
```

```
import nltk #this part is needed on colab.
nltk.download('punkt')
nltk.download('stopwords')
#----------------------------------------

# Tokenizing sentences
from nltk.tokenize import sent_tokenize

# Tokenizing words
from nltk.tokenize import word_tokenize

# Import stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
```

```
sentences = sent_tokenize(text)

# Use word_tokenize to tokenize the third sentence: tokenized_sent
tokenized_sent = word_tokenize(sentences[0])

# Make a set of unique tokens in the entire scene: unique_tokens
unique_tokens = set(word_tokenize(text))

print(tokenized_sent)
print(unique_tokens)
```

```
# A VERY UNELEGANT WAY OF DOING IT, BUT IT WORKS..........................
[[word.lower() for word in word_tokenize(sentences[0]) if word not in stop_words and word.isalnum()],[word.lower() for word in word_tokenize(sentences[1]) if word not in stop_words and word.isalnum()], [word.lower() for word in word_tokenize(sentences[2]) if word not in stop_words and word.isalnum()], [word.lower() for word in word_tokenize(sentences[3]) if word not in stop_words and word.isalnum()]]
```

```
import pandas as pd
pd.set_option('display.max_colwidth', -1) #to see more text

import numpy as np
import seaborn as sns

import itertools
from collections import Counter
```

```
# Tokenizing Tweets made easy!
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()
```

```
# download and open some Trump tweets from trump_tweet_data_archive

trump_tweets_df = pd.read_json('https://github.com/CALDISS-AAU/sdsphd19_coursematerials/raw/master/data/tweets_boomer.zip')
trump_tweets_df.head()
```

```
# Reset index (not really needed but why not)
trump_tweets_df = trump_tweets_df.set_index(pd.to_datetime(trump_tweets_df.created_at))
```

```
# Let's identify people Trump likes to mention
trump_tweets_df['mentions'] = trump_tweets_df['tweet'].map(lambda textline: [tag for tag in tknzr.tokenize(textline) if tag.startswith('#')])
```

```
# Only keep tweets where a mention i present
trump_tweets_df = trump_tweets_df[trump_tweets_df['mentions'].map(len) > 0]
```

```
# Collect (as the rows contain lists, the '*' opens up the lists, so it's a long list instead of a list of a list)
trump_tags = itertools.chain(*trump_tweets_df['mentions'])
```

```
# Count up and show
counted_tags = Counter(trump_tags)
counted_tags.most_common()[:10]
```

```
tknzr = TweetTokenizer(strip_handles=True)
trump_tweets_df['tokenized'] = trump_tweets_df.clean_tweet.map(lambda t: tknzr.tokenize(t))
```

```
trump_tweets_df['tokenized'] = trump_tweets_df['tokenized'].map(lambda t: [token.strip('@') for token in t if token.isalpha()])
```

```
# Import the dictionary builder
from gensim.corpora.dictionary import Dictionary
```

```
# Create a Dictionary from the articles: dictionary
dictionary = Dictionary(trump_tweets_df['tokenized'])
```

```
# filter out low-frequency / high-frequency stuff, also limit the vocabulary to max 1000 words
dictionary.filter_extremes(no_below=5, no_above=0.5, keep_n=1000)
```

```
# construct corpus using this dictionary
corpus = [dictionary.doc2bow(doc) for doc in trump_tweets_df['tokenized']]
```
